### Now we see that the "Diagnosis" column is categorical, and we are doing linear regression
### So, convert the categorical column to numerical as in 0 & 1.

#### Check for the various values in the categorical column ~ diagnosis

```python
breast_cancer_data.diagnosis.value_counts()
```

#### Use label encoder package to change the cat to numerical

```python
label_encode = LabelEncoder()
```

```python
label = label_encode.fit_transform(breast_cancer_data.diagnosis)
```

```python
breast_cancer_data['target'] = label
```

```python
breast_cancer_data.head(20)
```

```python
### Drop the original categorical column, 
```

```python
breast_cancer_data.drop(columns='diagnosis', axis = 1, inplace= True)
```

### Group the data based on the mean

```python
breast_cancer_data.groupby('target').mean()
```

## Inference: Mean values for Malignant is higher than Beging

# Summary of EDA
    1. The data had no missing value.
    2. All variables are numerical in nature.
    3. Mean is higher than the median for most of the variables, so the data is right skewed.
    4. Slight imbalance in the dataset: Beging cases are more than the Malignant cases.
    5. Mean of most the variables are larger for Malignant cases compared to the Beging cases
